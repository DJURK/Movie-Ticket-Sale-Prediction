# SCRAPING RATING DATA FROM IMDB

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import pandas as pd
import re

import os
chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver


driver = webdriver.Chrome(chromedriver)
```

```python
#read in previously scraped data
df = pd.read_csv('movies1.csv')
```

```python
df.head()
```

```python
#Make sure all the titles are correct
df1 = df['movie title']
for row in df1:
    print (row)
```

```python
#Initializing an empty list
IMDB_list = []
```

```python
#Selenium finds the search bar, clicks, sends a movie name, then scrapes the rating from the page.
for row in df1:
    try:
        url = "http://www.imdb.com"
        driver.get(url)
        query = driver.find_element_by_id("navbar-query")
        query.send_keys(row)
        query.send_keys(Keys.RETURN)
        time.sleep(1)

        name_selector = '//*[@id="main"]/div/div[2]/table/tbody/tr[1]/td[2]/a'
        driver.find_element_by_xpath(name_selector).click()
        rating = driver.find_element_by_xpath('//span[@itemprop="ratingValue"]').text

        headers = ['movie title', 'IMDB rating']

        time.sleep(1)
        imdb_dict = dict(zip(headers, [row, rating]))
        print (imdb_dict)
        IMDB_list.append(imdb_dict)
        
    except:
        pass
    
```

```python
len(IMDB_list)
```

```python
pd.DataFrame(IMDB_list).to_csv('IMDBfinal.csv')
```
